```python
dados1['order_purchase_timestamp'] = pd.to_datetime(dados1['order_purchase_timestamp']) # Transformei os dados desta coluna que estava em Str em datetime, posteriormente dividi o mês em duas partes, pode vir a trazer um insight interessante ou não. 
dados1['dias'] = dados1['order_purchase_timestamp'].dt.day  
def categorizar_dia(dia):
    if dia <= 15:
        return 'primeiros 15 dias'
    else:
        return 'ultimos dias'
dados1['dias_do_mes'] = dados1['dias'].apply(categorizar_dia)
```

```python
def funcao_vendas_por_mes(dados1, coluna_data='order_purchase_timestamp', coluna_id='order_id'):
    dados1[coluna_data] = pd.to_datetime(dados1[coluna_data]) # Convertendo a coluna para datetime
    dados1['mes_vigente'] = dados1[coluna_data].dt.to_period('M').dt.to_timestamp()  # Criando uma nova coluna mes vigente que recebe os dados com o método que agrupa as datas em meses
    vendas_por_mes = dados1.groupby('mes_vigente')[coluna_id].count().reset_index(name='contagem_de_vendas') # Agrupa os dados pelo mês e conta os numeros de pedidos
    return vendas_por_mes
```
